### Aplicando trasformaçoes: filtrando registros da categoria livros

```python
collection.distinct('Categoria do Produto')
```

```python
# importando pandas e criando um dataframe com os dados da coleção filtrando apenas a categoria livros
import pandas as pd
query = {'Categoria do Produto': 'livros'}

lista_livros = []
for doc in collection.find(query):
    lista_livros.append(doc)

df_livros = pd.DataFrame(lista_livros)
df_livros.head()
```

## Formatando as Datas

```python
df_livros["Data da Compra"] = pd.to_datetime(df_livros["Data da Compra"])
df_livros.info()
```

# Aplicando trasformações: filtrando produtos vendidos a partir de 2021

Utilizando uma Regex(expressão regular) =  sequência de caracteres para definir um padrão

```python
collection.find_one()
```

```python
query = {'Data da Compra': {'$regex': '/202[1-9]'}}

produtos = []
for doc in collection.find(query):
    produtos.append(doc)

df_produtos = pd.DataFrame(produtos)
df_produtos.head()
```

```python
df_produtos['Data da Compra'] = pd.to_datetime(df_produtos['Data da Compra'], format= '%d/%m/%Y')
df_produtos['Data da Compra'] = df_produtos['Data da Compra'].dt.strftime('%Y-%m-%d')
df_produtos.head()
```
